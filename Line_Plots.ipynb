````python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
````

````python
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv')
df.head()
````

````python
df.set_index('Country', inplace=True)
df.index.name = None
df.head()
````

````python
years = list(map(str, range(1980, 2014)))
````

````python
haiti = df.loc['Haiti', years]
haiti
````

````python
haiti.plot(kind = 'line')
plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')
plt.show()
````

````python
haiti.index = haiti.index.map(int)
haiti.plot(kind = 'line')
plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')
plt.text(2000, 6000, '2010 Earthquake')
plt.show()
````

````python
df_CI = df.loc[['India','China'], years]
df_CI.head()
````

````python
df_CI.plot(kind = 'line')
````

````python
df_CI = df_CI.transpose()
df_CI.head()
````

````python
df_CI.index = df_CI.index.map(int)
df_CI.plot(kind = 'line')
plt.title('Immigration from India and China')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')
plt.show()
````
